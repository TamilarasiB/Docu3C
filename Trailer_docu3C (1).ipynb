```python
import os
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv ("/content/complete-data.csv")
```

```python
dataset
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
dataset.groupby('emotion').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(dataset['emotion'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(dataset, x='frame_number', y='emotion', inner='box', palette='Dark2')
sns.violinplot(title='Emotion vs Year')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
data_woman=dataset[dataset["gender"]=="woman"]
data_man=dataset[dataset["gender"]=="man"]
```

```python
data_woman
```

```python
from matplotlib import pyplot as plt
data_woman['year'].plot(kind='hist', bins=20, title='year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data_man
```

```python
from matplotlib import pyplot as plt
data_man['year'].plot(kind='hist', bins=20, title='year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
woman_frames = data_woman.groupby(by=['year'])['frame_number'].count().to_frame(name = 'count').reset_index()
man_frames = data_man.groupby(by=['year'])['frame_number'].count().to_frame(name = 'count').reset_index()
```
