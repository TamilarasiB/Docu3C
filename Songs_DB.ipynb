```python
from matplotlib import pyplot as plt
dataset[' FEMALE SINGER COUNT'].plot(kind='hist', bins=20, title=' FEMALE SINGER COUNT')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
dataset[' MALE SINGER COUNT'].plot(kind='hist', bins=20, title=' MALE SINGER COUNT')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
dataset.rename(columns={"YEAR":"Year","MALE SINGER COUNT":"Male_Singer","FEMALE SINGER COUNT":"Female_Singer","MALE SONG COUNT":"Male_Song", "FEMALE SONG COUNT":"Female_Song"},inplace=True)
```

```python
dataset["TOTAL SINGERS"] = dataset[' FEMALE SINGER COUNT'] + dataset[" MALE SINGER COUNT"]
```

```python
dataset["TOTAL SONGS"] = dataset[' FEMALE SONG COUNT'] + dataset[" MALE SONG COUNT"]
```

```python
dataset
```

```python
from matplotlib import pyplot as plt
dataset.plot(kind='scatter', x=' YEAR', y=' MALE SINGER COUNT', s=32, alpha=.8, title='Male Singer Count')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
dataset.plot(kind='scatter', x=' YEAR', y=' FEMALE SINGER COUNT', s=32, alpha=.8, title='Female Singer Count')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
dataset.plot(kind='scatter', x=' MALE SONG COUNT', y=' FEMALE SONG COUNT', s=32, alpha=.8, title='Male Song vs Female Song')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
